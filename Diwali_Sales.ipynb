```python
ax = sns.countplot(data = df, x = 'gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.countplot(data = df, x = 'age_group')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_by_gender = df.groupby('gender', as_index=False)['amount'].sum()

ax = sns.barplot(x = 'gender', y = 'amount', data = sales_by_gender)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.countplot(x = 'age_group', data = df, hue = 'gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby('age_group', as_index = False)['amount'].sum().sort_values(by = 'amount', ascending = False)

ax = sns.barplot(x = 'age_group', y = 'amount', data = df.groupby('age_group', as_index = False)['amount'].sum().sort_values(by = 'amount', ascending = False))

# for bars in ax.containers:
#     ax.bar_label(bars)
```

```python
orders_by_state = df.groupby('state', as_index = False)['orders'].sum().sort_values(by = 'orders', ascending = False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
ax = sns.barplot(x = 'state', y = 'orders', data = orders_by_state)

for bars in ax.containers:
    ax.bar_label(bars)

df.groupby('state', as_index = False)['orders'].sum().sort_values(by = 'orders', ascending = False).head(10)
```

```python
# total amount/sales from top 10 states

df.groupby('state', as_index = False)['amount'].sum().sort_values(by = 'amount', ascending = False)

sales_by_states = df.groupby('state', as_index = False)['amount'].sum().sort_values(by = 'amount', ascending = False).head(10)

ax = sns.barplot(x = 'state', y = 'amount', data = sales_by_states)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.countplot(x = 'marital_status', data = df, hue = 'gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.countplot(x = 'occupation', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_by_occupation = df.groupby('occupation', as_index = False)['amount'].sum().sort_values(by = 'amount', ascending = False)

sns.barplot(data = sales_by_occupation, x = 'occupation', y = 'amount' )
```
